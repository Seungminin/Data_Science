```python
wt = np.random.uniform(low=40, high=90, size=100)
ht = np.random.uniform(low=1.4, high=2, size = 100)
```

```python
def BMI(wt,ht):
    bmi = wt/(ht*ht)
    return bmi
```

```python
bmi_arr = BMI(wt,ht)
print(bmi_arr)
```

```python
bmi = pd.DataFrame(bmi_arr)
print(bmi.head(10))
```

```python
status = ['Underweight', 'Healthy', 'Overweight', 'Obese']
count = np.zeros((4,1))

for num in bmi_arr:
    if num < 18.5:
        count[0]+=1
    elif 18.5<=num<=24.9:
        count[1]+=1
    elif 25<=num<=29.9:
        count[2]+=1
    elif num>=30:
        count[3]+=1

print(count.shape)
```

```python
plt.bar(status,count[:,0])
```

```python
plt.hist(bmi_arr,bins=4, edgecolor ='black')
```

```python
plt.pie(count[:,0],labels=status, autopct="%1.2f%%")
```
