```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"column_a" : [1,2,4,4,np.nan,np.nan,6],
                   "column_b" : [1.2,1.4,np.nan,6.2,None,1.1,4.3],
                   "column_c" : ['a','?','c','d','--',np.nan,'6'],
                   "column_d" : [True,True,np.nan,None,False,True,False]})

#nan, None은 float 형태의 Null값이다.
```

```python
df
```

```python
new = pd.Series([1, 2, np.nan, 4, np.nan, 5],dtype = pd.Int64Dtype())
df['column_e'] = new
```

```python
df

#NA는 integer Null값이다.
```

```python
df.isna() #df.isna() = df.isnull()은 같은 표현이다. -> df.notna는 opposite개념, 반대 개념이다. false를 true로 출력을 하는.
```

```python
df.isna().any() #어떤 행의 NUll값이 있냐?
```

```python
df.isna().sum() #Missing value가 몇개 있느냐?
```

```python
'''missing_values = ["?","--"] #?와 --를 missing_value로 처리하는 방법이다.
df_test = pd.read_csv("dataset.csv", na_values = missing_values)'''

df.replace({'?': np.nan, '--':np.nan}, inplace=True) #?, -- 표현을 Nan missing data로 변환, inplace parameter는 변환을 저장하는 역할
#inplace defalut값은 false이다.
df
```

```python
df.dropna(axis = 0, how = 'all', inplace = True) #dropna는 drop하는 method이고 axis = 0은 row행을 삭제하는 것, 
#how = all은 모두 missing data이면 row행을 삭제, how의 default는 any이다.
#axis = 1이면 열을 삭제
df
```

```python
df.dropna(axis=0, thresh=3,inplace=True) #Thresh는 missing data의 개수를 나타낸다. thresh이상이면 삭제
df
```
