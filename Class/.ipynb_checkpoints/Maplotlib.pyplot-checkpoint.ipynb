```python
import numpy as np
import matplotlib.pyplot as plt
wt = np.random.uniform(low = 40,high = 90, size = 100)

ht = np.random.uniform(low =1.4, high = 2.0, size = 100)

BMI = wt/ht**2
BMI_arr = np.array(BMI)
BMI_arr = np.round(BMI_arr,2)

print(BMI_arr)
```

```python
status = ['Underweight', 'Healthy', 'Overweight', 'Obese'] #status
status_indices = [0]*4 #index for the number of status
data = [[] for _ in range(len(status))] #각 status의 해당하는 data들의 모음

for a in BMI_arr:
    if a < 18.5:
        status_indices[0]+=1
        data[0].append(a)
    elif 18.5 <= a <=24.9:
        status_indices[1]+=1
        data[1].append(a)
    elif 25 <= a <=29.9:
        status_indices[2]+=1
        data[2].append(a)
    else:
        status_indices[3]+=1
        data[3].append(a)

print(status_indices)
plt.figure(figsize=(4,3))
plt.title("BMI distribution")
plt.xlabel("Status")
plt.ylabel("Distribution")
plt.bar(status,status_indices)
plt.show()
```

```python
plt.pie(status_indices,labels = status, autopct = '%1.2f%%')
plt.title("BMI distribution")
plt.show()
```

```python
plt.hist(BMI_arr,bins = [0,18.5,24.9,29.9,100])
plt.title("histogram of result")
plt.xticks([0,18.5,24.9,29.9,100])
plt.show()
```
