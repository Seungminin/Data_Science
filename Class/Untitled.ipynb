```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm #3 정규분포 만들기
```

```python
test_score = [28,35,26,32,28,28,35,34,46,42,37]
test = np.array(test_score)
```

```python
sum(test)
```

```python
mean(test)
```

```python
mean = np.mean(test)
```

```python
std_dev = np.std(test)
```

```python
print(5.98621*11)
```

```python
for i in test:
    z = (i-33)/6
    print(z, end = " ")
```

```python
x = np.linspace(0, 70, 1000)
normal_dist = norm.pdf(x, mean, std_dev)

# 데이터 히스토그램 그리기
plt.hist(test, bins=10, density=True, alpha=0.6, color='g', label='Histogram')

# 정규 분포 곡선 그리기
plt.plot(x, normal_dist, 'r--', label='Normal Distribution')

# 라벨 및 타이틀 추가
plt.title('Histogram of Test Scores with Normal Distribution')
plt.xlabel('Test Scores')
plt.ylabel('Probability Density')
plt.legend()

# 그래프 보이기
plt.show()
```
