```python
from sklearn.cluster import KMeans

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("C:/Users/ksmin/data_Science/dataset/clustering_datasets.csv", header=None)
dataset.columns=['height','weight']
print(dataset.head())
print()
print(dataset.isna().sum()) #There is no NaN value.
print()
print(dataset.info()) #2 rows and 1100 columns.
```

```python
plt.title("Distribution dataset")
plt.scatter(dataset['height'], dataset['weight'], c='blue')
plt.xlabel("height")
plt.ylabel('weight')
plt.grid()
plt.show()
```

```python
#Create 2 clusters, change max_iter : 50,100,200,300
Kmeans = KMeans(n_clusters=2, max_iter=200).fit(dataset)
Centroids = Kmeans.cluster_centers_
print(Centroids)

colors = ['blue', 'green']
# Display the 2 clusters with different labels for each group
for i in range(2):
    plt.scatter(dataset['height'][Kmeans.labels_==i],dataset['weight'][Kmeans.labels_==i], c = colors[i], s = 50, alpha=0.5, label = f'Group {chr(65+i)}')
plt.grid()
plt.scatter(Centroids[:,0], Centroids[:,1], color='red', s=50, label='centroid')
plt.legend(loc = 'lower right')
plt.show()
```
