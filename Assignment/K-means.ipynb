```python
from sklearn.cluster import KMeans

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = {'A' : [1.0, 1.5, 3.0, 5.0, 3.5, 4.5, 3.5],
        'B' : [1.0,2.0,4.0,7.0,5.0,5.0,4.5]}

data = pd.DataFrame(dataset)

print(len(data['A']))
print(data.head(7))
```

```python
#Create 3 clusters
Kmeans = KMeans(n_clusters=3).fit(data)
Centroids = Kmeans.cluster_centers_
print(Centroids)

# Display the 2 clusters with different labels for each group
plt.scatter(data['A'], data['B'], c=np.where(Kmeans.labels_==0, 'blue', 'green'), s=50, alpha=0.5, label='Group')
plt.grid()
plt.scatter(Centroids[:,0], Centroids[:,1], color='red', s=50, label='centroid')
plt.legend()
plt.show()
```
