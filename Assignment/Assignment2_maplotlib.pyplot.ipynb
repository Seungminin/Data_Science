```python
#Code
import numpy as np
import matplotlib.pyplot as plt
wt = np.random.uniform(low = 40,high = 90, size = 100)

ht = np.random.uniform(low =1.4, high = 2.0, size = 100)

BMI = wt/ht**2
BMI_arr = np.array(BMI)
BMI_arr = np.round(BMI_arr,2)

print(BMI_arr)
#########################################
#Bar Chart
status = ['Underweight', 'Healthy', 'Overweight', 'Obese'] # 4 status
status_indices = [0]*4 #The number of each status

for a in BMI_arr:
    if a < 18.5:
        status_indices[0]+=1
    elif 18.5 <= a <=24.9:
        status_indices[1]+=1
    elif 25 <= a <=29.9:
        status_indices[2]+=1
    else:
        status_indices[3]+=1

print(status_indices)
plt.figure(figsize=(4,3))
plt.title("BMI distribution")
plt.xlabel("Status")
plt.ylabel("Distribution")
plt.bar(status,status_indices)
plt.show()
#########################################
#histogram
plt.hist(BMI_arr, bins=[0,18.5,24.9,29.9,40], edgecolor = 'black'). #range  A~B ,bins=N   .
plt.title("histogram of BMI distribution.")
plt.show()
#########################################
#Pie chart
plt.pie(status_indices,labels = status, autopct = '%1.2f%%')
plt.title("BMI distribution")
plt.show()
#########################################
#Scatter Plot
w_h_range = np.linspace(40,210,100)
plt.scatter(w_h_range, wt, color='r', label='Weight')  # x,y  . data_type, data  .
plt.scatter(w_h_range, ht*100, color='b', label='Height')
plt.title("Scatter Plot")
plt.xlabel("Range")
plt.ylabel("Value")
plt.legend()
plt.show()
```
